```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
def monte_carlo_pi(runs):
    in_circle = 0
    for i in range(runs):
        x, y = np.random.uniform(0, 1, size=2)
        r = (x**2 + y**2)**0.5
        in_circle = in_circle + 1*(r < 1)
    return in_circle/runs*4
```

```python
sample = np.logspace(1, 4, num=100) #ilość próbek

approximations = [monte_carlo_pi(int(i))
                     for i in sample] #wyznaczanie wartości aproksymacji Pi

error = [np.abs(i-(3.14159265))
         for i in approximations] #wartość bezwględna (aproksymacja Pi - wartość Pi=3.14159265)
```

```python
plt.figure(figsize=(20, 10))
plt.title("Aproksymacja stałej π metodą Monte Carlo", fontsize=15) 
plt.xlabel("Rozmiar próbki", fontsize=15) 
plt.ylabel("Błąd aproksymacji", fontsize=15) 
plt.plot(sample, error) 
plt.show()
```
